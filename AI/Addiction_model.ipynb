# Feature Engineering

```python
df['Check_Per_Hour']=df['Phone_Checks_Per_Day']/(df['Daily_Usage_Hours']+1)
df['App_Switching_Rate']=df['Apps_Used_Daily']/(df['Daily_Usage_Hours']+1)
df['Usage_Sleep_Ratio']=df['Daily_Usage_Hours']/(df['Sleep_Hours']+1)
df["nonproductive_ratio"] = (df["Time_on_Social_Media"] + df["Time_on_Gaming"]) / (df["Time_on_Education"] + 1)
#df["mental_stress_index"] = (df["Anxiety_Level"] + df["Depression_Level"]) / (df["Self_Esteem"] + 1)
df["focus_ratio"] = df["Time_on_Education"] / (df["Time_on_Social_Media"] + 1)
df["bedtime_ratio"] = df["Screen_Time_Before_Bed"] / (df["Sleep_Hours"] + 1)
df["sleep_disturbance_score"] = df["Screen_Time_Before_Bed"] * df["Phone_Checks_Per_Day"] / (df["Sleep_Hours"] + 1)
```

# Rule-based Scoring (Brainrot)

```python
import pandas as pd
import numpy as np

# =========================================
# 1. قائمة الخصائص (Features)
# =========================================
# تم تصحيح الكوما المفقودة وترتيب العواميد
Brainrot_features = [
    'Time_on_Social_Media', 'nonproductive_ratio', 'App_Switching_Rate',
    'Phone_Checks_Per_Day', 'Check_Per_Hour', 'Daily_Usage_Hours',
    'Apps_Used_Daily', 'Time_on_Gaming', 'Screen_Time_Before_Bed',
    'Sleep_Hours', 'sleep_disturbance_score', 'Anxiety_Level',
    'Social_Interactions', 'focus_ratio'
]

# عمل نسخة للعمل عليها (نفترض أن df موجود ومحمل بالبيانات)
df_scaled = df[Brainrot_features].copy()

# توحيد الاتجاه: (كل ما الرقم زاد = برين روت أكتر)
df_scaled['Sleep_Hours'] = 1 - df_scaled['Sleep_Hours']
df_scaled['Social_Interactions'] = 1 - df_scaled['Social_Interactions']
df_scaled['focus_ratio'] = 1 - df_scaled['focus_ratio']

# =========================================
# 2. حساب المجموعات (Weighted Groups)
# =========================================
# الأوزان مدروسة بحيث يكون المجموع الكلي 1.0 (100%)

overstimulation = (
    0.15 * df_scaled['Time_on_Social_Media'] +
    0.10 * df_scaled['Apps_Used_Daily'] +
    0.05 * df_scaled['Time_on_Gaming']
)

addiction_behavior = (
    0.10 * df_scaled['Phone_Checks_Per_Day'] +
    0.05 * df_scaled['Check_Per_Hour'] +
    0.10 * df_scaled['App_Switching_Rate'] 
)

sleep_damage = (
    0.10 * df_scaled['Screen_Time_Before_Bed'] +
    0.05 * df_scaled['Sleep_Hours'] + 
    0.05 * df_scaled['sleep_disturbance_score']
)

lifestyle = (
    0.05 * df_scaled['Daily_Usage_Hours'] +
    0.05 * df_scaled['nonproductive_ratio'] +
    0.05 * df_scaled['focus_ratio']
)

mental_social = (
    0.05 * df_scaled['Anxiety_Level'] +
    0.05 * df_scaled['Social_Interactions']
)

# حساب السكور النهائي لكل صف
df['Brainrot_Score'] = overstimulation + addiction_behavior + sleep_damage + lifestyle + mental_social

# =========================================
# 3. تحديد المسطرة (Logical Thresholds)
# =========================================
# هنحسب المتوسط والانحراف عشان نحدد "المنطقي"
mean_score = df['Brainrot_Score'].mean()
std_score = df['Brainrot_Score'].std()

# المسطرة المنطقية:
# Low: أي حد تحت (المتوسط - نصف انحراف معاري) -> يعني شخص أهدى من الطبيعي
# High: أي حد فوق (المتوسط + نصف انحراف معاري) -> يعني شخص "سحله" أعلى من الطبيعي
low_limit = mean_score - (1 * std_score)
high_limit = mean_score + (1 * std_score)

def classify_logic(score):
    if score < low_limit:
        return "Low (Mild)"
    elif score > high_limit:
        return "High (Severe)"
    else:
        return "Medium (Moderate)"

# تصنيف كل صف بناءً على السكور بتاعه بالنسبة للعمود
df['Brainrot_Stage'] = df['Brainrot_Score'].apply(classify_logic)

# =========================================
# 4. النتائج النهائية
# =========================================
print(f"--- Logical Thresholds Used ---")
print(f"Low Limit: {low_limit:.3f} | High Limit: {high_limit:.3f}\n")

print("--- Distribution of Stages ---")
print(df['Brainrot_Stage'].value_counts())

print("\n--- Sample of Highest Brainrot Scores ---")
print(df[['Brainrot_Score', 'Brainrot_Stage']].sort_values(by='Brainrot_Score', ascending=False).head(5))
```

```python
import pandas as pd
import numpy as np

# =========================================
# 1. استخراج الـ Feature Importances
# =========================================
if hasattr(best_xgb, 'feature_importances_'):
    importances = best_xgb.feature_importances_
    feature_names = X_train.columns
    model_importance_map = dict(zip(feature_names, importances))
else:
    # fallback في حالة عدم وجود موديل جاهز
    model_importance_map = {}

# =========================================
# 2. قاموس التوصيات الشامل (لكل Feature)
# =========================================
# تأكد أن الأسماء هنا تطابق تماماً أسماء الأعمدة في X_train
feature_recommendations = {
    "Time_on_Social_Media": "Limit social media usage to specific time blocks (e.g., 30 mins post-work).",
    "App_Switching_Rate": "Practice 'Deep Work'; focus on one app at a time and close background tasks.",
    "Sleep_Hours": "Your sleep duration is low. Try to get at least 7-8 hours of consistent sleep.",
    "Screen_Time_Before_Bed": "Stop using your phone at least 60 minutes before sleep to improve melatonin levels.",
    "Phone_Checks_Per_Day": "Break the habit of compulsive checking. Keep your phone in another room while working.",
    "Daily_Usage_Hours": "Your total screen time is high. Set a hard limit using 'Digital Wellbeing' tools.",
    "Anxiety_Level": "High digital anxiety detected. Try mindfulness apps or the 4-7-8 breathing technique.",
    "Check_Per_Hour": "Frequent interruptions detected. Turn off non-essential notifications.",
    "Social_Interactions": "Your digital life is overshadowing real-world social cues. Plan an offline meet-up.",
    "nonproductive_ratio": "A large portion of your time is spent on non-productive apps. Audit your app list."
}

# =========================================
# 3. النظام المعتمد على الـ Features فقط
# =========================================
def feature_based_recommendation_system(user, importance_map):
    contributions = {}
    
    # حساب تأثير كل Feature عند المستخدم ده
    for feature, weight in importance_map.items():
        if feature in user:
            # القيمة الفعلية للمستخدم * وزن الميزة في الموديل
            # ملاحظة: لو الميزة عكسية (زي النوم)، بنستخدم (1 - value) لو البيانات Normalized
            val = user[feature]
            
            # معالجة الميزات العكسية (كل ما قلت كان التأثير السلبي أكبر)
            if feature in ['Sleep_Hours', 'Social_Interactions']:
                val = 1 - val 
                
            contributions[feature] = weight * val

    # ترتيب الميزات حسب الأعلى تأثيراً (Highest Contribution to the problem)
    sorted_factors = sorted(contributions.items(), key=lambda x: x[1], reverse=True)
    
    total_impact = sum(contributions.values())
    recs = []
    top_factors_report = []

    # اختيار أهم 3 ميزات مسببة للمشكلة
    for f_name, impact_val in sorted_factors[:3]:
        percent = round((impact_val / total_impact) * 100, 1) if total_impact > 0 else 0
        
        # إضافة التقرير الرقمي
        top_factors_report.append({
            "feature": f_name,
            "impact_weight": percent
        })
        
        # إضافة النصيحة المرتبطة بالميزة
        if f_name in feature_recommendations:
            recs.append(feature_recommendations[f_name])

    return {
        "Analysis_Status": "Based on Feature Importance",
        "Top_Contributing_Features": top_factors_report,
        "Personalized_Recommendations": list(dict.fromkeys(recs)) # منع التكرار
    }

# =========================================
# 4. التجربة
# =========================================
# user_sample = X_test.iloc[0] # مثال لمستخدم
# result = feature_based_recommendation_system(user_sample, model_importance_map)
# print(result)
```
